```python
import cv2
import json
from ultralytics import YOLO

model_path = 'models/yolov8/detect/train2/weights/best.pt'
model = YOLO(model_path)
```

```python
cap = cv2.VideoCapture('rtsp://192.168.1.8:8080/h264_aac.sdp')
# cap = cv2.VideoCapture(1)
skip_factor = 20
count = 0
conf_thres = 0.75
font = cv2.FONT_HERSHEY_SIMPLEX
import time

while True:
    # print(frame.shape)
    # time.sleep(1)
    ret, frame = cap.read()
    if ret:
        if count % skip_factor == 0:  
            result = model(frame)
            predictions = json.loads(result[0].tojson())
            for prediction in predictions:
                conf_score = prediction['confidence']

                if conf_score < conf_thres:
                    continue
                    
                bbox = prediction['box']
                name = prediction['name']
                xmin = int(bbox['x1'])
                ymin = int(bbox['y1'])
                xmax = int(bbox['x2'])
                ymax = int(bbox['y2'])
                font_size = 0.7


                cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0, 255, 0), 3)

                text =name + "-conf: " + f"{conf_score:.2f}"
                
                (text_width, text_height), _ = cv2.getTextSize(text, font, font_size, 2)

                cv2.rectangle(frame, (xmin, ymin - text_height - 5), (xmin + text_width, ymin), (0,255,0), -1)
                cv2.putText(frame, text, (xmin, ymin -5), font, font_size, (0,0,0),2)

            cv2.imshow('frame', frame)
        count += 1


        
        if cv2.waitKey(1) == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
